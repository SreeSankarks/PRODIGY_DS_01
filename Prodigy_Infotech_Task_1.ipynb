```python
# data for total population
total_population_data = data[data["Series Code"] == "SP.POP.TOTL"]

# data based on the total population for 2022
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)

# top fifteen countries with the highest total population for 2022
total_fifteen_countries = total_population_sorted.head(15)
print("Top fifteen countries of total population")
print(total_fifteen_countries )
```

```python
top_15 = total_population_sorted.head(15)
bottom_15 = total_population_sorted.tail(15)

# Plotting
sns.set(style="whitegrid")
plt.figure(figsize=(14, 12))

# Top 15 countries bar plot
plt.subplot(2, 1, 1)
sns.barplot(x='2022', y='Country Name', data=top_15, palette='pastel')
plt.title('Top 15 Countries by Population in 2022')
plt.xlabel('Population')
plt.ylabel('Country')

# Bottom 15 countries bar plot
plt.subplot(2, 1, 2)
sns.barplot(x='2022', y='Country Name', data=bottom_15, palette='pastel')
plt.title('Bottom 15 Countries by Population in 2022')
plt.xlabel('Population')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

To create stacked bar charts for the top and bottom 15 countries based on male and female population data for the year 2022,

```python
top_fifteen_male = male_population_sorted.head(15)
bottom_fifteen_male = male_population_sorted.tail(15)

top_fifteen_female = female_population_sorted.head(15)
bottom_fifteen_female = female_population_sorted.tail(15)

# Function to plot stacked bar chart
def plot_stacked_bar(data_male, data_female, title):
    countries = data_male['Country Name']
    male_population_2022 = data_male['2022']
    female_population_2022 = data_female['2022']
    
    # Plotting
    plt.figure(figsize=(10, 6))
    
    # Plot bars for male population
    plt.bar(countries, male_population_2022, label='Male Population', color='b')
    
    # Plot bars for female population stacked on top of male population
    plt.bar(countries, female_population_2022, bottom=male_population_2022, label='Female Population', color='g')
    
    # Customize the plot
    plt.xlabel('Countries')
    plt.ylabel('Population')
    plt.title(title)
    plt.xticks(rotation=90)
    plt.legend()
    
    # Show plot
    plt.tight_layout()
    plt.show()

# Plot stacked bar chart for top 15 countries (male and female)
plot_stacked_bar(top_fifteen_male, top_fifteen_female, 'Top 15 Countries Male vs Female Population (2022)')

# Plot stacked bar chart for bottom 15 countries (male and female)
plot_stacked_bar(bottom_fifteen_male, bottom_fifteen_female, 'Bottom 15 Countries Male vs Female Population (2022)')
```
